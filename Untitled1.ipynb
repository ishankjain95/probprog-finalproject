# Autoregressive Models
An autoregressive model is when time series values (our stock prices) are regressed on previous values from that same time series data. For example, if a day's stock price was regressed on yesterday's stock price we would get
$$ y_t = \beta_0 + \beta_1 y_{t-1} + \epsilon_t $$
In autoregressive models the response variable in the previous time period has become the predictor in the current time period. The error term follows the same structure as a normal linear regression. Since we are only using the previous time period as our predictor this is an AR(1) model. Below we use an AR(2) model, meaning we use the previous two days of stock prices as predictors for the current stock price. Our formula becomes 
$$ y_t = \beta_0 + \beta_1 y_{t-1} + \beta_2 y_{t-2} + \epsilon_t $$
We can then generalize to an AR(k) model which uses the previous k time periods as predictors in the current time period. The formula becomes: 
$$ y_t = \beta_0 + \sum_{i=1}^k \beta_i y_{t-i} + \epsilon_t $$


```python
# Parameters

mu = 0.
beta_true = np.array([0.7, 0.25])

noise_obs = 0.1
#T = 128
p = 2

#ed.set_seed(42)

data = pd.read_csv('data/Google.csv')
data = data['Close']
data = data.as_matrix()
x_true = np.zeros(128)
for i in range(128):
    x_true[127-i] = data[data.size-1-i]
plt.plot(x_true)
plt.show()

## Format data as percent change
for i in reversed(range(1,x_true.size)):
    x_true[i] = round((x_true[i]-x_true[i-1])/x_true[i-1],2)
x_true[0] = 0
N = 128
plt.plot(x_true)
plt.show()
print("Number of data points: {}".format(N))
T = N
```

```python
from edward.models import Normal, InverseGamma, PointMass

mu = Normal(loc=0.0, scale=10.0)
beta = [Normal(loc=0.0, scale=2.0) for i in range(p)]

noise_proc = tf.constant(0.1) #InverseGamma(alpha=1.0, beta=1.0)
noise_obs = tf.constant(0.1) #InverseGamma(alpha=1.0, beta=1.0)

x = [0] * T
for n in range(p):
    x[n] = Normal(loc=mu, scale=10.0)  # fat prior on x
for n in range(p, T):
    mu_ = mu
    for j in range(p):
        mu_ += beta[j] * x[n-j-1]
    x[n] = Normal(loc=mu_, scale=noise_proc)

# use variance as zero to convert the list of x's to something Inference
# will understand
#y = Normal(mu=x, sigma=tf.constant(0.))
```

## MAP & KL
Below we perform MAP and KL inference to get the mu and betas for our autoregressive model. In addition, we used a python library to create a model to compare with. Similar to the Hidden Markov Model we can see that KL doesn't do as well at infering the mu and betas. This follows the same principals as we discussed in the Hidden Markov Model section as to the issues with KL on time series data. It does seem to do better on our autoregressive model compared to our hidden markov model. MAP does come with it's downfalls, one being that it is a point estimation and so you miss out on information about the distribution. However, our MAP estimate comes very close to the values from our python library. 


```python
print("setting up distributions")
qmu = PointMass(params=tf.Variable(0.))
qbeta = [PointMass(params=tf.Variable(0.)) for i in range(p)]
print("constructing inference object")
vdict = {mu: qmu}
vdict.update({b: qb for b, qb in zip(beta, qbeta)}) 
inference = ed.MAP(vdict, data={xt: xt_true for xt, xt_true in zip(x, x_true)})
print("running inference")
inference.run()

print("parameter estimates:")
print("beta: ", [qb.value().eval() for qb in qbeta])
print("mu: ", qmu.value().eval())
```

```python
print("setting up variational distributions")
qmu = Normal(loc=tf.Variable(0.), scale=tf.nn.softplus(tf.Variable(0.)))
qbeta = [Normal(loc=tf.Variable(0.), scale=tf.nn.softplus(tf.Variable(0.))) for i in range(p)]
print("constructing inference object")
vdict = {mu: qmu}
vdict.update({b: qb for b, qb in zip(beta, qbeta)})
inference_vb = ed.KLqp(vdict, data={xt: xt_true for xt, xt_true in zip(x, x_true)})
print("running inference")
inference_vb.run()

print("parameter estimates:")
for j in range(p):
    print("beta[%d]: " % j, qbeta[j].mean().eval(),)
    #" +- ", qbeta[j].scale().eval())
print("mu: ", qmu.variance().eval())
#, qmu.scale().eval())
```
